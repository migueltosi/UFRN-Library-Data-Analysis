```python
import pandas as pd
import numpy as np
```

```python
i = 20
year_filename = ['20201']
for n in range(0,10):
  i -= 1
  year_filename.append('20' + str(i) + '1')
  year_filename.append('20' + str(i) + '2')
```

```python
dataframes = []

for i in year_filename:
  year = ''.join(list(i)[:-1])

  df = pd.read_csv(f'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{i}.csv', sep=',')
  df['Ano'] = year

  dataframes.append(df)
```

```python
merged_df = pd.concat(dataframes).sort_values('Ano')
```

```python
loans = merged_df
```
